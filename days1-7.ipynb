```python
def get_rank(h):
    d_h = {}
    high_rank = 0
    #low_rank = 0
    for c in list(h):
        d_h[c] = 1 + d_h.get(c,0)
    
    
    if len(d_h)==1:
        high_rank = 8 #5 of a kind
        #low_rank = CARD_VALUES[d_h.keys()[0]]
    elif len(d_h)==2: #either full house or four of a kind
        v=list(d_h.values())
        v.sort()
        if v[0]==1:
            #four of a kind
            high_rank = 7
        else:
            #full house
            high_rank = 6
    elif len(d_h)==3:
        v=list(d_h.values())
        v.sort()
        if v[1]==1:
            high_rank = 5 #Three of a kind AAABC
        else:
            high_rank = 4 #Two pair AABBC
    elif len(d_h)==4:
        high_rank = 3 #One pair AABCD
    else:
        high_rank = 2 #High card ABCDE
    
    return high_rank
```

```python
def get_max_rank(h):
    mr=get_rank(h)
    #print(f'--{h}--')
    for ix,c in enumerate(list(h)):
        if c=='J':
            for nc in list(CARDS):
                newh=list(h)
                newh[ix]=nc
                newr=get_max_rank(newh)
                if newr > mr :
                    #print(f'>> {newh}')
                    mr = newr
    return mr
    
```

```python
from pathlib import Path
import itertools as it

def minloc(seeds: list[int], maps: list[list[tuple[int, int, int]]]):
    sol = seeds[0]
    for seed in seeds:
        for _map in maps:
            for dest, source, length in _map:
                if source <= seed < source + length:
                    seed += -source + dest
                    break
        sol = sol if sol < seed else seed
    return sol

def intervals(locs: list[range],  _map: list[tuple[int, int, int]]):
    next_locs = []
    for r in locs:
        stop = False
        for dest, source, length in _map:
            if source <= r.start < r.stop <= source + length:
                stop = True
                next_locs.append(range(r.start - source + dest, r.stop - source + dest))
                break
            if source < r.start < source + length <= r.stop:
                next_locs.append(range(r.start - source + dest, dest + length))
                next_locs.append(range(source + length, r.stop))
                break
            if r.start < source and source + length <= r.stop:
                next_locs.append(range(dest, dest + length))
                next_locs.append(range(source + length, r.stop))
                break
        if not stop:
            next_locs.append(r)
    return next_locs

def rangeloc(ranges, maps):
    allranges = []
    for _map in maps:
        ranges = intervals(ranges, _map)
        allranges.append(ranges)
    return min(map(lambda x: x.start, it.chain.from_iterable(allranges)))

def solutions():
    data = inp.split('\n\n')
    seeds = [int(s) for s in data[0].split(': ')[1].split(' ')]
    print((seeds))
    ranges = [range(seeds[i], seeds[i+1]) for i in range(int(len(seeds)/2))]
    print(ranges)
    maps = []
    sol1, sol2 = seeds[0], seeds[0]
    for m in data[1:]:  # For each mapping, put it in a list and the triplets in tuples
        tmp = []
        for vals in m.split(':')[1].strip().split('\n'):
            tmp.append(tuple(map(int, vals.split())))
        maps.append(sorted(tmp, key=lambda x: x[1])) # Sort on source start
    sol1 = minloc(seeds, maps)
    sol2 = rangeloc(ranges, maps)
    return sol1, sol2
                

solutions()
```

```python
import uuid

# import requests
# from requests.adapters import HTTPAdapter
from requests.packages.urllib3 import PoolManager
from requests.packages.urllib3.util.retry import Retry

import json


DEFAULT_TIMEOUT = 360

retry_strategy = Retry(
    total=3,
    status_forcelist=[429, 500, 502, 503, 504],
    method_whitelist=["HEAD", "GET", "POST", "OPTIONS"],
    backoff_factor=10
)
def mlapi_request(service, service_version, endpoint, body):
    request_id = str(uuid.uuid4())
     # currently only set up to do the POST requests
    assert(service in set(['sos', 'langid', 'jdparser', 'demographics', 'education']))  # nosec
    url = f'http://localhost:8000/v1.0/{service}/{service_version}/{endpoint}'
    r = None
    try:
        http = PoolManager(retries=retry_strategy)
        r = http.request('POST', url, body=json.dumps(body).encode('utf-8'), timeout=DEFAULT_TIMEOUT, headers={'Content-Type': 'application/json'})

        return json.loads(r.data.decode('utf-8'))
    except Exception as e:
        print(f'{request_id} API call to {url} failed hard {e}')
        return None
```

```python
def get_lang(txt):
    data = mlapi_request("langid",'v1.1', "predict", {
        'text': txt
    })
    #print(data)
    return data.get('lang','')
```
