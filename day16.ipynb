```python
def move(i,j, nexti,nextj):
    if arr[nexti][nextj] == '.':
        return [[nexti+(nexti-i),nextj+(nextj-j)]]

    if arr[nexti][nextj] == '|':
        #N-S
        if nextj==j:
            return [[nexti+(nexti-i),nextj]]
        #W-E split
        return [[nexti-1, nextj], [nexti+1, nextj]]
    elif arr[nexti][nextj] == '-':
        #W-E
        if abs(nextj-j)>0:
            return [[nexti,nextj+(nextj-j)]]
        #N-S split
        return [[nexti, nextj-1], [nexti, nextj+1]]
    elif arr[nexti][nextj] == '/':
        if nextj==j: # N,S
            if nexti>i: # from N
                return [[nexti, nextj-1]]
            else:
                return [[nexti, nextj+1]]
        else:
            if nextj>j:
                return [[nexti-1,nextj]]
            else:
                return [[nexti+1, nextj]]
    elif arr[nexti][nextj] == '\\':
        if nextj==j: # N,S
            if nexti>i: # from N
                return [[nexti, nextj+1]]
            else:
                return [[nexti, nextj-1]]
        else:
            if nextj>j:
                return [[nexti+1,nextj]]
            else:
                return [[nexti-1, nextj]]
        
            
```
