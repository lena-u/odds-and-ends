```python
def all_dots(l):
    return all([li=='.' for li in l])
```

```python
def expand_cols(arr):
    print(f'expand_cols starts')
    new_arr = copy.deepcopy(arr)
    k = list(range(len(arr[0])))
    k.reverse()
    for j in k:
        #print(i)
        if all_dots([arr[i][j] for i in range(len(arr)) ]):
            print(j, len(new_arr), len(new_arr[2]))
            [new_arr[i].insert(j, '.') for i in range(len(arr))]
            print(j, len(new_arr), len(new_arr[2]))
    return new_arr
```

```python
def expand_rows(arr):
    print(f'expand_rows starts')
    new_arr = copy.deepcopy(arr)
    k = list(range(len(arr)))
    k.reverse()
    for i in k:
        #print(i)
        if all_dots([arr[i][j] for j in range(len(arr[i])) ]):
            print(i, len(new_arr[i]), len(new_arr))
            new_arr.insert(i, ['.' for j in range(len(arr[i]))] )
            print(i, len(new_arr[i]), len(new_arr), len(new_arr[i]))

    return new_arr
```

```python
new_arr = expand_rows(expand_cols(arr))
```

```python
#manhattan distance
def dist(x,y):
    return abs(x[0]-y[0])+abs(x[1]-y[1])
```

```python
#get all galaxies
gals = {}
cnt = 1
for i in range(len(new_arr)):
    for j in range(len(new_arr[i])):
        if new_arr[i][j] == '#':
            gals[cnt] = [i,j]
            cnt = cnt+1
gals
```

```python
keys = gals.keys()
#print(keys)
n= len(keys)
d = []
#print("n",n)
for i in range(n):
    #print('from ',i+1,' to ',n)
    for j in range(i+1,n):
        #print("i",i, "j",j)
        d.append(dist(gals[i+1], gals[j+1]))
```
