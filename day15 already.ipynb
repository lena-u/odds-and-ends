```python
def ha(asc):
    #print(f'*** {asc}')
    h=0
    for a in asc:
        #print(a)
        if a=='-' or a=='=':
            break
        h = h+ord(a)
        h = h*17
        h = h % 256
    
    mybox = boxes[h]
    print(f'box {h} {mybox}')
    if '=' in asc:
        b = asc.split('=')
        #print('= ', b)  
        if b[0] in [x[0] for x in mybox]:
            #print('found it')
            ix = [x[0] for x in mybox].index(b[0])
            mybox[ix][1]=b[1]
        else:
            #print('adding it')
            mybox.append(b)
    elif '-' in asc:
        b = asc.split('-')
        #print('- ', b)  
        if b[0] in [x[0] for x in mybox]:
            #print('found it')
            ix = [x[0] for x in mybox].index(b[0])
            mybox.pop(ix)
    else:
        print('wtf', asc)
        pass
    
    #print('--- final :',mybox)
    
    return h
```
