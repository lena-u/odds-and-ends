```python
def find_v(p):
    c =-1
    pt = [['']*len(p) for i in range(len(p[0]))]

    for i in range(len(p)):
        for j in range(len(p[i])):
            #print(i,j)
            pt[j][i] = p[i][j]
    
    return find_h(pt)
    
```

```python
def find_h(p):
    c =-1
    cs = []
    for i in range(len(p)-1):
        #print(p[i],'vs\n',p[i+1])
        if p[i]==p[i+1]:
            cs.append(i)
            
    #check the rest
    flag = True
    #print('found cs',cs)
    if len(cs)==0:
        return -1
    for c in cs:
        flag=True
        for i in range(int(len(p)/2)):
            #print('compare',c-i,c+i+1,":",  p[c-i], p[c+i+1])
            if c-i<0 or c+i+1> len(p)-1:
                #print('!!')
                break
            if p[c-i]!=p[c+1+i]:
                #print('not eq')
                flag=False
                break
        #print('so', flag, c)
        if flag:
            return c
    return -1
```

```python
def alter_kaker(p,i,j):
    ap= copy.deepcopy(p)
    ap[i][j]='.' if ap[i][j]=='#' else '#'
    return (find_h(ap)+1, find_v(ap)+1)
```
